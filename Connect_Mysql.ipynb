```python
# Thêm thư viện mysql.connector vào chương trình
import mysql.connector
```

```python
# Thêm thư viện pandas vào chương trình
import pandas as pd
```

```python
# Tạo biến db để lưu database connect vào server
db= mysql.connector.connect(user='mindx_sql',password='mindx_sql', host='103.173.254.194', database='classicmodels', auth_plugin='mysql_native_password') 
```

```python
# Chạy query để lấy dữ liệu đưa vào biến data
mycursor = db.cursor()
mycursor.execute("SELECT * FROM employees")
data = mycursor.fetchall()
```

```python
# Chuyển data thành dataFrame
data= pd.DataFrame(data, columns=['Employeenumber','Lastname','Firstname','Extension','Email','Officecode','Reportsto','Jobtitle'])
```
